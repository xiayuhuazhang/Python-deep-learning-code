1. get Boston_housing data

```python
from keras.datasets import boston_housing
(train_data, train_targets), (test_data, test_targets) = \
    boston_housing.load_data()
print('train_data.shape:', train_data.shape)
print('test_data.shape:', test_data.shape)
print('train_targets:', train_targets)
```

2. nomalize the data

```python
mean = train_data.mean(axis=0)
train_data -= mean
std = train_data.std(axis=0)
train_data /= std
test_data -= mean
test_data /= std
```

3. define and compile the model

```python
from keras import models, layers
def build_model(): #functional model
    model = models.Sequential()
    model.add(layers.Dense(64, activation='relu', 
        input_shape=(train_data.shape[1], )))
    model.add(layers.Dense(64, activation='relu'))
    model.add(layers.Dense(1)) #scaler regression
    # loss-->(mean squared error)，
    # train-->(mean absolute error)
    model.compile(optimizer='rmsprop', loss='mse', metrics=['mae'])
    return model
```

4. K-fold to validate the performance

```python
import numpy as np 
k = 4
num_val_samples = len(train_data) // k
num_epochs = 100
all_scores = []
for i in range(k):
    print('processing fold #', i)
    # set val set
    val_data = train_data[i * num_val_samples: (i + 1) * num_val_samples]
    val_targets = \
        train_targets[i * num_val_samples: (i + 1) * num_val_samples]
    # set the trian set    
    partical_train_data = np.concatenate(
        (train_data[: i * num_val_samples], 
            train_data[(i + 1) * num_val_samples: ]), axis=0
    )
    partical_train_targets = np.concatenate(
        (train_targets[: i * num_val_samples], 
            train_targets[(i + 1) * num_val_samples: ]), axis=0
    )
    # instance a model
    model = build_model() 
    # train the model ,silent mode: verbose=0
    model.fit(partical_train_data, partical_train_targets, epochs=num_epochs, 
        batch_size=1, verbose=0)
    # validate the model
    val_mse, val_mae = model.evaluate(val_data, val_targets, verbose=0)
    all_scores.append(val_mae)
print('all_scores:', all_scores)
print('mean scores:', np.mean(all_scores))
```

5. maxmize the epoches and visualize

```python
import matplotlib.pyplot as plt 
num_epochs = 500
all_mae_history = []
for i in range(k):
    print('processing fold #', i)
    # set val set
    val_data = train_data[i * num_val_samples: (i + 1) * num_val_samples]
    val_targets = \
        train_targets[i * num_val_samples: (i + 1) * num_val_samples]
    # set the trian set    
    partical_train_data = np.concatenate(
        (train_data[: i * num_val_samples], 
            train_data[(i + 1) * num_val_samples: ]), axis=0
    )
    partical_train_targets = np.concatenate(
        (train_targets[: i * num_val_samples], 
            train_targets[(i + 1) * num_val_samples: ]), axis=0
    )

    model = build_model() #instance 1 model
    # train
    history = model.fit(partical_train_data, partical_train_targets, 
        validation_data=(val_data, val_targets), epochs=num_epochs, 
        batch_size=1, verbose=0)
    mae_history = history.history['val_mae']
    all_mae_history.append(mae_history)
# calculate the mean score of k-fold
average_mae_history = [np.mean([x[i] for x in all_mae_history]) for i in range(num_epochs)]
```

```python
plt.plot(range(1, len(average_mae_history) + 1), average_mae_history)
plt.xlabel('Epochs')
plt.ylabel('Validation MAE')
plt.show()
```

```python
# replot to get better result(delete 10 at begin, and smooth the curve)
def smooth_curve(points, factor=0.9):
    smoothed_points = []
    for point in points:
        if smoothed_points:
            previous = smoothed_points[-1]
            smoothed_points.append(previous * factor + point * (1 - factor))
        else:
            smoothed_points.append(point)
    return smoothed_points
smoothed_mae_history = smooth_curve(average_mae_history[10: ])
plt.plot(range(1, len(smoothed_mae_history) + 1), smoothed_mae_history)
plt.xlabel('Epochs')
plt.ylabel('Validation MAE')
plt.show()
```

6. train the final model

```python
# overfit after 80 epoches so set epoches to 80
# also can tune the hidden layer size
model = build_model()
model.fit(train_data, train_targets, epochs=80, batch_size=16, verbose=0)
test_mse_score, test_mae_score = model.evaluate(test_data, test_targets)
print('Test mae score:', test_mae_score)
```
